```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dates = pd.date_range('2023-01-01', '2024-11-14', freq='10min')
#dates
```

```python
#df = pd.Series(np.random.randn(len(dates)), dates, name="rand")
#df.plot()
#df
```

```python
df_len = len(dates)
step = 0.1
stop = df_len * step
df = pd.Series(
    np.sin(
        np.arange(start=0, stop=stop, step=step)
    ), dates, name="seq")
df
```

```python
close=df.resample("1h").ohlc()["close"].tail(200)
```

```python
fig, ax = plt.subplots()
ax.scatter(df.tail(120).index, df.tail(120), label="seq", color="b")
ax.scatter(close.tail(20).index, close.tail(20), label="close", color="r")
```

## Linear

```python
b = 0.1
w = 0.01

def line(x: np.array):
    return x * w + b

df_len = len(dates)
step = 0.1
stop = df_len * step

arg = np.arange(start=0, stop=stop, step=step)
y = line(arg)

df = pd.Series(y, dates, name="seq")
df
```

```python
close=df.resample("1h").ohlc()["close"].tail(200)

fig, ax = plt.subplots()
ax.scatter(df.tail(120).index, df.tail(120), label="seq", color="b")
ax.scatter(close.tail(20).index, close.tail(20), label="close", color="r")
```
